```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('ggplot')
seed = 101 # Lucky seed
np.random.seed = seed
```

```python
def target(x):
    return x*np.sin(x)

X = np.linspace(-2, 10, 1000)
y = target(X)
```

Grid Search figure.

```python
X_grid = np.linspace(-2, 10, 10)
y_grid = target(X_grid)

plt.figure(figsize=(6,3))
plt.plot(X, y, color='black', label='Target')
plt.scatter(X_grid, y_grid, color='tomato', zorder=20, label='Samples')
plt.legend(loc='best')
plt.title('Grid')
plt.xlabel('$x$')
plt.ylabel('$Target$')
plt.show()
```

Random Search figure.

```python
X_random = np.random.uniform(-2, 10, 10)
y_random = target(X_random)

plt.figure(figsize=(6,3))
plt.plot(X, y, color='black', label='Target')
plt.scatter(X_random, y_random, color='tomato', zorder=20, label='Samples')
plt.legend(loc='best')
plt.title('Random')
plt.xlabel('$x$')
plt.ylabel('$Target$')
plt.show()
```

Iterative subsampling figure.

```python
X_grid[y_grid.argsort()[::-1][0]]
```

```python
# Get position of previous max
X_max = X_grid[y_grid.argsort()[::-1][0]]
X_grid2 = np.linspace(X_max-2, X_max+2, 10)
y_grid2 = target(X_grid2)

plt.figure(figsize=(6,3))
plt.plot(X, y, color='black', label='Target')
plt.scatter(X_grid, y_grid, color='tomato', zorder=20, label='Stage 1')
plt.scatter(X_grid2, y_grid2, color='steelblue', zorder=20, label='Stage 2')
plt.axvspan(X_max-2, X_max+2, color='steelblue', alpha=0.5)
plt.legend(loc='best')
plt.title('Iterative Subsampling')
plt.xlabel('$x$')
plt.ylabel('$Target$')
plt.show()
```
